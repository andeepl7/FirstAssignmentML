1.1 Exploratory Data Analysis (EDA)

```python
import pandas as pd
import seaborn as sns
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/nick-edu/dmmldl/master/MobilePrice.csv")
```

```python
#Counting rows and columns with info
df.info
```

```python
#Counting columns
len(df.columns)
```

```python
#Counting rows
len(df.index)
```

```python
#Counting missing values in the dataframe
df.isnull().sum()
```

```python
#Visualizating data
sns.histplot(data=df, x="ram", hue="wifi",element="poly")
```

```python
sns.barplot(data=df, x="n_cores", y="battery_power", hue ="dual_sim", palette = "Blues")
```

```python
sns.countplot(x = "screen_height", hue = "four_g", data = df, palette = 'magma')
```
